# Import the library

```python
import matplotlib.pyplot as plt
from datasets import load_dataset
```

# Load the data

```python
data_path = "cnamuangtoun/resume-job-description-fit"
```

```python
datasets = load_dataset(data_path)
```

```python
train_dataset = datasets['train']
test_dataset = datasets['test']
```

## Convert to pandas dataframe

```python
train_df = train_dataset.to_pandas()
test_df = test_dataset.to_pandas()
```

```python
train_df.head()
```

## Show if the data is balanced

```python
lable_count = train_df.groupby('label').count()
lable_count
```

```python
labels = ["Good Fit","No Fit","Potential Fit"]

label_figer=plt.figure(figsize=(7,3),dpi=100)
label_axes=label_figer.add_axes([0.1,0.1,0.9,0.9])

wedges, texts, autotexts=label_axes.pie(lable_count["resume_text"].values,labels=labels,
                                      autopct="%1.0f%%")
plt.legend(wedges,labels,loc="right",bbox_to_anchor=(1,0,0.5,1))

plt.show()
```
